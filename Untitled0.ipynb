```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('fraudTest.csv')
```

```python
# Preprocessing
# Convert datetime columns
data['trans_date_trans_time'] = pd.to_datetime(data['trans_date_trans_time'])
data['dob'] = pd.to_datetime(data['dob'])
```

```python
# Calculate age
data['age'] = data['trans_date_trans_time'].dt.year - data['dob'].dt.year
```

```python
# Encode categorical variables
le = LabelEncoder()
data['gender'] = le.fit_transform(data['gender'])
data['city'] = le.fit_transform(data['city'])
```

```python
# Select features
selected_features = data[['amt', 'age', 'gender']]
target = data['is_fraud']
```

```python
X_train, X_test, y_train, y_test = train_test_split(selected_features, target, test_size=0.2, random_state=42, stratify=target)
```

```python
# Apply SMOTE for balancing
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)
```

```python
# Bar chart showing fraud count in each age group
plt.figure(figsize=(10, 6))
data['age_group'] = pd.cut(data['age'], bins=[0, 20, 40, 60, 80, 100], labels=['0-20', '21-40', '41-60', '61-80', '81-100'])
age_fraud_counts = data.groupby('age_group')['is_fraud'].sum()
sns.barplot(x=age_fraud_counts.index, y=age_fraud_counts.values, palette='viridis')
plt.title('Fraud Count in Each Age Group')
plt.xlabel('Age Group')
plt.ylabel('Fraud Count')
plt.show()
```

```python
# Initialize and train the Gradient Boosting Classifier
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gb_model.fit(X_train_balanced, y_train_balanced)

# Optionally, evaluate the model using cross-validation
gb_cv_scores = cross_val_score(gb_model, X_train_balanced, y_train_balanced, cv=5)
print("Gradient Boosting Cross-Validation Scores:", gb_cv_scores)
print(f"Gradient Boosting Mean Accuracy: {np.mean(gb_cv_scores):.2f}")
```
